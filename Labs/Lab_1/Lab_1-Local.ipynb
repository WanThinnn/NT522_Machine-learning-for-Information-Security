# **Predict Data**
Công cụ dự đoán các giá trị trong tương lai từ một tập nhiều đầu vào và trên trọng số tương đối, bằng cách ứng dụng tích giữa ma trận và vector

```python
import numpy as np
def predict(data, w):
    return data.dot(w)

# w is the vector of weights
w = np.array([0.1, 0.2, 0.3])

# matrices as input datasets
data1 = np.array([0.3, 1.5, 2.8])
data2 = np.array([0.5, 0.4, 0.9])
data3 = np.array([2.3, 3.1, 0.5])
data_in = np.array([data1[0],data2[0],data3[0]])
print('Predicted value: $%.2f' % predict(data_in, w) )
```

### Linear Regression với scikit-learn

Chúng ta sẽ dùng một bộ dữ liệu giả lập rất đơn giản:

Đầu vào: X = [1, 2, 3, 4, 5]

Đầu ra: y = [2, 4, 6, 8, 10] (tức y = 2x)

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Dữ liệu
X = np.array([[1], [2], [3], [4], [5]])  
y = np.array([2, 4, 6, 8, 10])

# Khởi tạo và huấn luyện mô hình
model = LinearRegression() # Khởi tạo mô hình
model.fit(X, y) # Huấn luyện mô hình

# Dự đoán
y_pred = model.predict(X) # Dự đoán giá trị y cho các điểm dữ liệu X
print("scikit-learn predictions:", y_pred) # In kết quả dự đoán


# Cảm nghĩ về scikit-learn:
# Ưu điểm: Giao diện của scikit-learn cực kỳ đơn giản, dễ sử dụng và phù hợp với các mô hình học máy truyền thống. Nó cho phép ta nhanh chóng xây dựng, huấn luyện và đánh giá các mô hình mà không cần quá nhiều cấu hình phức tạp.
#Nhược điểm: Thư viện này không hỗ trợ các thuật toán deep learning hiện đại và có giới hạn khi xử lý các bài toán phi tuyến phức tạp hoặc vấn đề với dữ liệu lớn và đa dạng. Khi cần xây dựng các mô hình như mạng nơ-ron sâu, ta sẽ cần đến các thư viện chuyên biệt như TensorFlow hoặc PyTorch.

```

```python
import numpy as np
import tensorflow as tf

# Dữ liệu
X = np.array([[1], [2], [3], [4], [5]], dtype=np.float32)
y = np.array([2, 4, 6, 8, 10], dtype=np.float32)

# Khởi tạo các biến cho trọng số và bias
W = tf.Variable([[1.0]], dtype=tf.float32)
b = tf.Variable(0.0, dtype=tf.float32)

# Hàm mô hình: y_pred = X*W + b
def model(X):
    return tf.matmul(X, W) + b

# Hàm mất mát: Mean Squared Error
def loss_fn(y_true, y_pred):
    return tf.reduce_mean(tf.square(y_true - y_pred))

optimizer = tf.optimizers.SGD(learning_rate=0.01)

# Huấn luyện mô hình với vòng lặp
epochs = 1000
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        y_pred = model(X)
        loss = loss_fn(y, y_pred)
    gradients = tape.gradient(loss, [W, b])
    optimizer.apply_gradients(zip(gradients, [W, b]))
    if epoch % 100 == 0: # In loss sau mỗi 100 epochs
        print(f"Epoch {epoch}: loss = {loss.numpy()}")

print("TensorFlow predictions:", model(X).numpy())

# Cảm nghĩ về TensorFlow:
# Ưu điểm: Mạnh mẽ, có khả năng mở rộng cao và hỗ trợ tính toán phân tán, GPU; cho phép tùy biến ở mức thấp.
# Nhược điểm: Giao diện API cấp thấp có thể khá phức tạp đối với người mới bắt đầu.
```

### Linear Regression với Keras (API cấp cao trên TensorFlow)

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# Dữ liệu
X = np.array([[1], [2], [3], [4], [5]], dtype=np.float32)
y = np.array([2, 4, 6, 8, 10], dtype=np.float32)

# Xây dựng mô hình đơn giản với 1 lớp Dense
model = keras.Sequential([
    layers.Dense(1, input_shape=(1,))
])

# Biên dịch mô hình
model.compile(optimizer='sgd', loss='mean_squared_error')

# Huấn luyện mô hình
model.fit(X, y, epochs=1000, verbose=0)

y_pred = model.predict(X)
print("Keras predictions:", y_pred)

# Cảm nghĩ về Keras:
# Ưu điểm: API rất thân thiện, dễ sử dụng, giúp nhanh chóng xây dựng và thử nghiệm mô hình deep learning.
# Nhược điểm: Do là API cấp cao, có thể giới hạn một số tùy biến sâu nếu bạn cần kiểm soát tối đa quá trình huấn luyện.
```

### Linear Regression với PyTorch

```python
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim

# Dữ liệu
X = np.array([[1], [2], [3], [4], [5]], dtype=np.float32)
y = np.array([[2], [4], [6], [8], [10]], dtype=np.float32)

# Chuyển đổi sang tensor
X_tensor = torch.from_numpy(X)
y_tensor = torch.from_numpy(y)

# Định nghĩa mô hình: một lớp Linear
model = nn.Linear(1, 1)

criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

epochs = 1000
for epoch in range(epochs):
    optimizer.zero_grad()
    outputs = model(X_tensor)
    loss = criterion(outputs, y_tensor)
    loss.backward()
    optimizer.step()
    if epoch % 100 == 0:
        print(f"Epoch {epoch}: loss = {loss.item()}")

predictions = model(X_tensor).detach().numpy()
print("PyTorch predictions:", predictions)

# Cảm nghĩ về PyTorch:
# Ưu điểm: API trực quan, linh hoạt, được ưa chuộng trong cộng đồng nghiên cứu; hỗ trợ tự động hóa tính toán gradient rất tốt.
# Nhược điểm: Cần viết nhiều code hơn so với Keras cho một số tác vụ, mặc dù nó lại cho phép kiểm soát chi tiết hơn quá trình huấn luyện.
```

# **Spam Detection with Perceptrons**

```python
import pandas as pd
import numpy as np
df = pd.read_csv('./resources/sms_spam_perceptron.csv')
y = df.iloc[:, 0].values
y = np.where(y == 'spam', -1, 1)
X = df.iloc[:, [1, 2]].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron(max_iter=40, eta0=0.1, random_state=0)
p.fit(X_train, y_train)
```

```python
y_pred = p.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print('Misclassified samples: %d' % (y_test != y_pred).sum())
print('Accuracy: %.2f' % accuracy_score(y_test, y_pred))
```
